# Import Libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sb
# Importing the dataset
dataset = pd.read_csv('quality.csv')
df=pd.DataFrame(dataset)
```

### Correlation

```python
#correlation using graph
plt.figure(figsize=(16,8))
sb.heatmap(df.corr(), annot=True)
plt.show()
```

### Add X and y Variable from dataset here X(Depenedent Variables) and y(Independent Variable)

```python
print(df.keys())
#X=df[['MemberID','ERVisits']]
#print(X)
X=df[['MemberID','ERVisits','OfficeVisits','Narcotics','DaysSinceLastERVisit','TotalVisits','ProviderCount','StartedOnCombination','AcuteDrugGapSmall']]
#X = dataset.iloc[:, 1:9].values
y = dataset.iloc[:, 13].values
print(X)
print(y)
print(X.shape)
```

#### Add Label to Dataset

```python
#print(X[:,-2])

from sklearn import preprocessing
labelencoder_X_1 = preprocessing.LabelEncoder()
#X[:, -2] = labelencoder_X_1.fit_transform(X[:, -2])
#print(X)
```

```python
print(X.shape)
```

#### Split data into training and testing in our case 30% testing and 70% training

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
print(X_train.shape)
print(X_test.shape)
print(y_test.shape)
print(y_train.shape)
```

#### Import Tensorflow, keras and Classifier

```python
import tensorflow as tf
import keras
from keras.models import Sequential
classifier=Sequential()
```

#### Add Cllassifier, Input Layer, Hidden Layer

```python
from keras.layers import Dense
from keras import losses

# Adding the input layer and the first hidden layer
classifier.add(Dense(10,input_dim = 9, kernel_initializer = 'uniform', activation = 'relu'))
# Adding the second hidden layer
classifier.add(Dense(10, kernel_initializer = 'uniform', activation = 'relu'))
# Adding the output layer
classifier.add(Dense(1, kernel_initializer = 'uniform', activation ='sigmoid'))

# Compile model
classifier.compile(loss='binary_crossentropy',optimizer='adam', metrics=['accuracy'])
# Fit the model
results=classifier.fit(X_train, y_train, epochs=150, batch_size=10, validation_data=(X_test, y_test))
```

#### Make Prediction

```python
# Predicting the Test set results
y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)
```

#### Confusion Matrix

```python
# Creating the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```
